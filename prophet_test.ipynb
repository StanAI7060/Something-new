```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# this program was generated by GPT4

# Example data
data = {
    'ds': pd.date_range(start='2022-01-01', periods=100),
    'y': (pd.Series(range(100)) + pd.Series(range(100)) * 0.5)  # Just as an example
}
df = pd.DataFrame(data)

# Initialize the model
model = Prophet()

# Fit the model
model.fit(df)

# Create a DataFrame for future predictions (e.g., for the next 30 days)
future_dates = model.make_future_dataframe(periods=30)

# Predict future values
forecast = model.predict(future_dates)

# Plot the forecast
fig = model.plot(forecast)
plt.title('Forecast of the Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()

# Plot the forecast components
fig2 = model.plot_components(forecast)
plt.show()
```
