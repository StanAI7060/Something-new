```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sine_wave(amplitude, frequency, duration):
    # Time array at 8000 samples per second
    t = np.linspace(0, duration, int(8000 * duration), endpoint=False)
    # Generate sine wave
    wave = amplitude * np.sin(2 * np.pi * frequency * t)
    return t, wave

def plot_sine_wave(t, wave):
    plt.figure(figsize=(10, 4))
    plt.plot(t, wave, color='red')  # Plotting the wave in red color
    plt.title('Sine Wave')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Predefined reasonable values
    amplitude = 1.0  # Sine wave amplitude
    frequency = 2.0  # Frequency in Hz
    duration = 2.0   # Duration in seconds

    # Generate and plot the wave
    t, wave = generate_sine_wave(amplitude, frequency, duration)
    plot_sine_wave(t, wave)
```
